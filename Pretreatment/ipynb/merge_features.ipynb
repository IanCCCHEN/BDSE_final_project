```python
#沒有category_repeat_buy_probability.csv
df_catrebuy=pd.read_csv('../train/category_repeat_buy_probability.csv')
df_productid = pd.read_csv('../train/productid.csv')
df_offers = pd.read_csv('../data/offers.csv')
df_userfeatures= pd.read_csv('../train/user_features.csv')
df_prodfeature=pd.read_csv('../train/product_features.csv')
df_seasonal=pd.read_csv('../train/seasonal_features.csv')
df_cheapness=pd.read_csv('../train/product_cheapness_features.csv')
df_rebuy_probability=pd.read_csv('../train/product_repeat_buy_probability.csv')
df_competition=pd.read_csv('../train/competition_features.csv')
df_new_prodfeature = pd.read_csv('../train/new_product_features.csv')
df_first_trans = pd.read_csv('../train/first_transaction_features.csv')
df_negative = pd.read_csv('../train/negative_features.csv')
df_base=pd.read_csv('../train/base_features.csv')
```

```python
df_catrebuy
```

```python
df_base
```

```python
data = pd.merge(df_base, df_userfeatures, on='id')
data
```

```python
data1 = pd.merge(data, df_seasonal, on='id')
data1
```

```python
data2 = pd.merge(data1, df_new_prodfeature, on='id')
data2
```

```python
data3 = pd.merge(data2, df_first_trans, on='id')
data3
```

```python
data4 = pd.merge(data3, df_negative, on='id')
data4
```

```python
data5 = pd.merge(data4, df_productid, left_on='offer_id',right_on='offer')
data5
```

```python
data6 = pd.merge(data5,df_prodfeature, on='productid')
data6
```

```python
data7 = pd.merge(data6,df_cheapness, left_on='productid',right_on='product_id')
data7
```

```python
data8 = pd.merge(data7,df_rebuy_probability, on='product_id')
data8
```

```python
data9 = pd.merge(data8,df_competition, left_on='category',right_on='categoryid')
data9
```

```python
data10 = pd.merge(data9,df_catrebuy, left_on='category',right_on='category_id')
data10
```

```python
#data11=data10.drop('product_id','category_id','categoryid',axis=1)
data11=data10.drop('product_id',axis=1)
data11=data11.drop('category_id',axis=1)
data11=data11.drop('categoryid',axis=1)
data11=data11.drop('offer',axis=1)
data11=data11.drop('offervalue',axis=1)
data11=data11.drop('quantity',axis=1)
data11=data11.drop('category',axis=1)
data11=data11.drop('company',axis=1)
data11=data11.drop('brand',axis=1)
data11
```

```python
data11.to_csv('../train/all_features.csv', encoding='big5', index=False)
```
