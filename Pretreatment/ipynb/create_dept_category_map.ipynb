```python
#製作catergory-dept對照表
import csv,time

# 開始測量
start = time.time()

with open("../data/transactions.csv","r") as fi:
    cat_dept_map = {}
    index = 0
    for lines in fi:
        index += 1
        li = lines.strip().split(",")
        cat_dept_map[li[3]] = li[2]
        if index % 5000000 == 0:
            print (index)

with open("../data/cat_dept_map.csv","w",newline='') as of:
    writer=csv.writer(of)
    writer.writerow(['catergory','dept'])
    for k,v in cat_dept_map.items():
        writer.writerow([k,v])
        
# 結束測量
end = time.time()

# 輸出結果
print("執行時間：%f 秒" % (end - start))
```
