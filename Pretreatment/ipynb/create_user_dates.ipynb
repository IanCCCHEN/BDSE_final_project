```python
# get first and last date of all users
import time

# 開始測量
start = time.time()

of = open("../data/user_dates.csv","w")
of.write("id,first_date,last_date\n")

fi = open("../data/transactions.csv","r")
fi.readline()
li = fi.readline().strip().split(",")
lastid = li[0]
startdate = li[6]
prevdate = li[6]
for i,lines in enumerate(fi):
	li = lines.strip().split(",")
	if not lastid == li[0]:
		# write out
		of.write(lastid+","+startdate+","+prevdate+"\n")
		# initalize new
		startdate = li[6]
	prevdate = li[6]
	lastid = li[0]
	if i % 5000000 == 0 and i > 0:
		print(i)
# last entry
of.write(lastid+","+startdate+","+prevdate+"\n")
fi.close()
of.close()

# 結束測量
end = time.time()

# 輸出結果
print("執行時間：%f 秒" % (end - start))
```
